Test GSEA Visualization

```python
from react_gsea import dataFromResult
from react_gsea.jupyter_compat import ReactGSEA
import numpy as np
import pandas as pd
import random
```

```python
with open('ag_up.txt') as upfile:
    upgenes = [line.strip() for line in upfile.readlines()]
    
with open('ag_down.txt') as downfile:
    downgenes = [line.strip() for line in downfile.readlines()]

print("Number of up-regulated genes:", len(upgenes))
print("Number of down-regulated genes:", len(downgenes))
print("First 5 up-regulated genes:", upgenes[:5])
print("First 5 down-regulated genes:", downgenes[:5])
```

```python
entlist = upgenes[:20] + downgenes[:20]
random.shuffle(entlist)

ranklist = list(range(10))
random.shuffle(ranklist)

ReactGSEA(
  data=dataFromResult({
    'up': upgenes[:10],
    'down': downgenes[:10],
  }, {
    'entities': entlist,
    'ranks': ranklist,
  }),
)
```
